```python
qPCR_file="data/Rn_cycle_curve.csv"

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data=pd.read_csv(qPCR_file,index_col=0)

A1=data.loc["A1",]
A2=data.loc["A2",]
A3=data.loc["A3",]
Hs_2_treat__C=(A1+A2+A3)/3


A4=data.loc["A4",]
A5=data.loc["A5",]
A6=data.loc["A6",]
Hs_2_treat__acC=(A4+A5+A6)/3

B1=data.loc["B1",]
B2=data.loc["B2",]
B3=data.loc["B3",]
Hs_2_control__C=(B1+B2+B3)/3


B4=data.loc["B4",]
B5=data.loc["B5",]
B6=data.loc["B6",]
Hs_2_control__acC=(B4+B5+B6)/3


plt.plot(Hs_2_treat__C)
plt.plot(Hs_2_treat__acC)

plt.plot(Hs_2_control__C)
plt.plot(Hs_2_control__acC)
```

```python
from plotnine import *
import pandas as pd
import numpy as np
from plotnine.data import mpg
import plotnine

plotnine.options.base_family = 'Arial'

black = '#222222'
gray = '#666666'
red = '#FF3333'
green = '#66CC00'
blue = '#3333FF'
purple = '#9933FF'
orange = '#FF8000'
yellow = '#FFFF33'
c1="#F8766D"
c2="#00BA38" 
c3="#619CFF"


cycle=[]
fluorescence_intencity=[]
site=[]

cycle.extend(list(np.linspace(1,40,40)))
fluorescence_intencity.extend(list(Hs_2_treat__C))
site.extend(["Control C site"]*40)

cycle.extend(list(np.linspace(1,40,40)))
fluorescence_intencity.extend(list(Hs_2_treat__acC))
site.extend(["Candidate ac4C site"]*40)


data=pd.DataFrame({"Cycle":cycle,"Fluorescence intencity":fluorescence_intencity,"site":site})

p=(ggplot()
 + geom_line(data,aes(x="Cycle",y="Fluorescence intencity",group="site",color="site"))
 + geom_point(data,aes(x="Cycle",y="Fluorescence intencity",group="site",color="site"),size=0.2, stroke=0.4,shape='s')
 #+ geom_text(aes(label="value"), size=10)
 +theme(panel_background=element_rect(fill=gray, alpha=0),
            panel_grid_major=element_line(size=0.3, alpha=0,color=black),
            panel_grid_minor=element_line(size=0.3, alpha=0,color=black),
            
            axis_text=element_text(size=6,family="Arial",color="black"),
            axis_title_x=element_text(size=6,family="Arial",color="black"),
            axis_title_y=element_text(size=6,family="Arial",color="black"),
            #legend_position =  "none",
            legend_title = element_text(size=6), #change legend title font size
            legend_text = element_text(size=6),
            legend_background=element_rect(size=0.5,alpha=0),
            legend_key=element_rect(size=0.5,alpha=0),
            legend_key_size=4,
            plot_title=element_text(margin={'b': 1, 'r': 0, 'units': 'pt'},size=6,family="Arial",color="black"),
            legend_position = (0.4,0.75),
            figure_size=[2,1.3],
            #panel_border=element_rect(color=black, size=1),
            axis_line_x=element_line(),  # Customize axis lines
            axis_line_y=element_line(),  # Customize axis lines
            panel_border=element_blank()
  )  
  #+scale_fill_gradient(low = "white", high = "#F8766D",limits=[0, 1]) 
+scale_color_manual([c1,c3])
+guides(color=guide_legend(title='Transcript NM_000224.3'))
+ggtitle("NaBH$_4$ treatment")
)

print(p)
p.save("figure/qPCR_NM_000224.3_NaBH4_treatment.pdf")
```
